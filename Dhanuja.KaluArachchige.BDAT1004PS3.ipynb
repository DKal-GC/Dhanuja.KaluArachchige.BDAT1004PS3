### Step 1,2,3

```python
import pandas as pd

users = pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
users.head()
```

### Step 4

```python
users.groupby('occupation')['age'].mean()
```

### Step 5

```python
users['male'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0)

user_male = users[users['gender'] == 'M']
male_count =  user_male.groupby('occupation', as_index=False)['user_id'].agg({'male_count':'count'})

total_count = users.groupby('occupation', as_index=False)['user_id'].agg({'user_count':'count'})

count_df = pd.merge(male_count,total_count, left_on='occupation',right_on='occupation')

count_df['male_ratio'] = count_df['male_count'] / count_df['user_count']
count_df.sort_values(by = 'male_ratio', ascending=False)
```

### Step 6

```python
occupation_ages = users.groupby('occupation', as_index=False).agg({'age':['min','max']})
occupation_ages
```

### Step 7

```python
pd.pivot_table(users, 'age','occupation','gender' ,aggfunc='mean')
```

### Step 8

```python
count_df['male_%'] = count_df['male_ratio'] * 100
count_df['female_%'] = 100 - count_df['male_%']
count_df[['occupation', 'male_%', 'female_%']]
```
